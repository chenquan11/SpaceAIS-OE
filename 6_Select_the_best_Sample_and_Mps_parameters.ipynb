```python
# dat = np.load('distance_cog_latlng_41108.npz')
dat = np.load('distance_cog_latlng.npz')
```

```python
distance_matrix = dat['distance_matrix']
lat=dat['lat']
lng=dat['lng']
cog=dat['cog']
```

```python
cog_matrix = dat['cog_matrix']
```

```python
# print(cog_matrix)
X = pd.DataFrame(columns=['lat','lng'])
# df_loss['epoch']=gru_epoch

X['lat']=lat
X['lng']=lng
```

```python

res = []
for eps in np.arange(1,181,5):

    for min_samples in range(4,33,4):#
#         dbscan = DBSCAN(eps = eps, min_samples = min_samples, metric='precomputed')
        dbscan = DBSCAN(eps = eps, min_samples = min_samples)
#         y_db = dbscan.fit_predict(distance_matrix)
        y_db = dbscan.fit_predict(cog_matrix)
        n_clusters = len([i for i in set(dbscan.labels_) if i != -1])
        # 异常点的个数
        outliners = np.sum(np.where(dbscan.labels_ == -1, 1,0))
        # 统计每个簇的样本个数
        stats = str(pd.Series([i for i in dbscan.labels_ if i != -1]).value_counts().values)
        res.append({'eps':eps,'min_samples':min_samples,'n_clusters':n_clusters,'outliners':outliners,'stats':stats})
        print("eps:",eps,"min_samples:",min_samples,"number:",len(stats))
#         print("n_clusters",(y_db))
        X['cluster'] = y_db 


df = pd.DataFrame(res)
```
